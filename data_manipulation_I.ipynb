```python
import pandas as pd
```

```python
df = pd.read_csv('sales_subset.csv')
df.head()
```

```python
print(df.info())
```

```python
# Subset for type A stores, calc total weekly sales
sales_A = df[df["type"] == "A"]["weekly_sales"].sum()
print(sales_A)
```

```python
# Subset for type B stores, calc total weekly sales
sales_B = df[df['type'] == 'B']['weekly_sales'].sum()
print(sales_B)
```

```python
# Subset for type C stores, calc total weekly sales
sales_C = df[df['type'] == 'C']['weekly_sales'].sum()
print(sales_C)
```

```python
# Get proportion for each type
sales_all = df["weekly_sales"].sum()
sales_proportion = [sales_A, sales_B, sales_C] / sales_all
print(sales_proportion)
```

```python
# Group by type; calc total weekly sales
sales_by_type = df.groupby("type")["weekly_sales"].sum()
print(sales_by_type)
# Get proportion for each type
sales_proportion = sales_by_type / sum(sales_by_type)
print(sales_proportion)
```

```python
# Group by type and is_holiday; calc total weekly sales
sales_by_type_is_holiday = df.groupby(["type", 'is_holiday'])["weekly_sales"].sum()
print(sales_by_type_is_holiday)
```

```python
# For each store type, aggregate weekly_sales: get min, max, mean, and median
import numpy as np
sales_stats = df.groupby('type')['weekly_sales'].agg([np.min, np.max, np.mean, np.median])
print(sales_stats)
```

```python
# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
unemp_fuel_stats = df.groupby('type')[['unemployment', 'fuel_price_usd_per_l']].agg([np.min, np.max, np.mean, np.median])
print(unemp_fuel_stats)
```
